#### edge detection

```python
box_img = np.zeros((400, 400))     # create blank greyscale img of 0s, each zero is balck color

box_img[70:90, 70:90] = 1          # This line sets a specific region within the image (from row 70 to 89 and column 70 to 89) 
# to have pixel values of 1. This effectively creates a white square in the image.


box_img = ndimage.rotate(box_img, 45, mode = 'constant')       # 45 degree rotation
# mode" parameter specifies how to handle areas outside the original image after rotation. 
# "constant" mode fills the missing areas with a constant value, which is 0 in this case.

box_img = ndimage.gaussian_filter(box_img, sigma = 5)

plt.imshow(box_img)
```

#### check if the small template is present in large one or no

```python
method = cv2.TM_SQDIFF_NORMED
```

```python
small_img = cv2.imread("G:\\DS150423\\DATA WRANGLING\\small.png")
results = []

large_img = cv2.imread("G:\\DS150423\\DATA WRANGLING\\large.png")
result = cv2.matchTemplate(small_img, large_img, method)
results.append(result)

threshold = 0.8
if result.any() >= threshold:
    print("found")
```
